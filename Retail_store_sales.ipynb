```python
import pandas as pd
import numpy as np

df = pd.read_csv("Retail_store_sales.csv", encoding='latin1')

df.head()
```

```python
df.info()
df.head()
```

```python
df.columns =df.columns.str.strip()
df.columns =df.columns.str.replace(' ','_')
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('\xa0', '')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Day'] = df['Order_Date'].dt.day
```

```python
df['Profit_Ratio'] = df['Profit'] / df['Sales']
```

```python
df.info()
```

```python
df.describe()
```

## Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", context="talk")  # clean + readable
plt.rcParams['figure.figsize'] = (9,5)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
```

#### Distribution Analysis

```python
plt.figure()
sns.histplot(df['Sales'], bins=30, kde=True)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.histplot(df['Quantity'], bins=20, kde=True)

plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure()
sns.barplot(x=cat_sales.index, y=cat_sales.values)

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
reg_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

plt.figure()
sns.barplot(x=reg_profit.index, y=reg_profit.values)

plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.tight_layout()
plt.show()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

```python
df['Order_Date'].dtype
```

```python
df['Month_Name'] = df['Order_Date'].dt.month_name()
```

```python
month_sales = df.groupby('Month_Name')['Sales'].sum()

month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

month_sales = month_sales.reindex(month_order)

plt.figure(figsize=(9,5))
sns.lineplot(x=month_sales.index, y=month_sales.values, marker='o')

plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.show()
```

```python
top_sub = df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure()
sns.barplot(x=top_sub.values, y=top_sub.index)

plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Sales")
plt.ylabel("Sub-Category")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()
```

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Category")
plt.show()
```

```python
df.groupby('Segment')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Segment")
plt.show()
```

```python
df.groupby('Category')['Sales'].sum().plot(
    kind='bar',
    color='skyblue'
)
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Sales by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Region")
plt.show()
```
